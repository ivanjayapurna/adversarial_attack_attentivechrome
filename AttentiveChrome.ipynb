# Results Visualisation & Analysis

```python
result_list = []
dir_list = []
for dir in sorted(os.listdir('./')):
  if 'Results' in dir:
    try:
      result_list.append(np.load(os.path.join(dir, 'E128total_auc.npy')))
      dir_list.append(dir)
    except Exception as e:
      print(e)

dir_list
```

```python
result_list = np.array(result_list)
print(result_list.shape)
result_list
```

```python
cell_lines = list(os.listdir('./data/'))
cell_lines.remove('E059')
cell_lines =  np.sort(cell_lines)
# truncate to only use E109 onwards (13 cell lines)
#cell_lines = cell_lines[42:]
cell_lines
```

```python
barWidth = 0.60
gap = 3
plt.figure(figsize=(70, 10))
colors = ['red', 'blue', 'orange', 'green', 'pink']
for i, results in enumerate(result_list):
  r = np.arange(len(results)) * gap + i * barWidth
  plt.bar(r, results, color=colors[i], width=barWidth, edgecolor='white', label=dir_list[i], linewidth=[1 for _ in results])

plt.title('AUCs of different attacks on all cell lines')
plt.xlabel('Cell lines')
plt.ylabel('AUC')
plt.xticks([r * gap + barWidth for r in range(len(results))], cell_lines)
plt.legend(loc='lower right')
plt.savefig('full_results.png')
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.title('AUCs of different attacks averaged across all cell lines')
plt.xlabel('Attacks')
plt.ylabel('AUC')
r = np.arange(len(result_list))
plt.xticks(r, labels=['' for _ in dir_list])
plt.bar(r, np.mean(result_list, axis=1), color=colors)
#plt.xticks(r, labels=dir_list)
legend_elements = [matplotlib.patches.Patch(facecolor=colors[i], label=dir_list[i]) for i in range(len(dir_list))]
plt.legend(handles=legend_elements, loc='lower right')
plt.savefig('average_results.png')
plt.show()
```

### Focus on specific cell lines
Look closer in depth to the best, worst and median performing cell lines.
First, we identify them.

```python
full_effectiveness = result_list[0]-result_list[1]
background_effectiveness = result_list[0]-result_list[2]
peak_effectiveness = result_list[0]-result_list[3]
print("full attack effectiveness", full_effectiveness)
print("non-peak region attack effectiveness", background_effectiveness)
print("peak region attack effectiveness", peak_effectiveness)
```

```python
attack_effectivenesses = [full_effectiveness, background_effectiveness, peak_effectiveness]
atk_eff_labels = ["full attack", "non-peak region attack", "peak region attack"]

barWidth = 0.60
gap = 3
plt.figure(figsize=(70, 10))
colors = ['red', 'blue', 'orange']
for i, results in enumerate(attack_effectivenesses):
  r = np.arange(len(results)) * gap + i * barWidth
  plt.bar(r, results, color=colors[i], width=barWidth, edgecolor='white', label=atk_eff_labels[i], linewidth=[1 for _ in results])

plt.title('Attack effectiveness of different attacks on all cell lines')
plt.xlabel('Cell lines')
plt.ylabel('AUC reduction from baseline')
plt.xticks([r * gap + barWidth for r in range(len(results))], cell_lines)
plt.legend(loc='lower right')
plt.savefig('attack_effectiveness.png')
plt.show()
```

```python
# Calculte Pearson's correlation between 3 attacks
def pearsonr(x, y):
  x_mean = np.mean(x)
  y_mean = np.mean(y)
  return np.sum((x - x_mean) * (y - y_mean)) / np.sqrt(np.sum((x - x_mean)**2) * np.sum((y - y_mean)**2))

print(pearsonr(full_effectiveness, background_effectiveness))
print(pearsonr(full_effectiveness, peak_effectiveness))
print(pearsonr(background_effectiveness, peak_effectiveness))
print(pearsonr(result_list[0], result_list[1]))
print(pearsonr(result_list[0], result_list[2]))
print(pearsonr(result_list[0], result_list[3]))
print(pearsonr(result_list[1], result_list[2]))
print(pearsonr(result_list[1], result_list[3]))
print(pearsonr(result_list[2], result_list[3]))
```

```python
# Choose some interesting cell lines to focus on:

# least successful attacks
print("Least successful attack:", cell_lines[np.argmin(full_effectiveness)])
print("Least successful background attack:", cell_lines[np.argmin(background_effectiveness)])
print("Least successful peak attack:", cell_lines[np.argmin(peak_effectiveness)])

# most successful attacks
print("Most successful attack:", cell_lines[np.argmax(full_effectiveness)])
print("Most successful background attack:", cell_lines[np.argmax(background_effectiveness)])
print("Most successful peak attack:", cell_lines[np.argmax(peak_effectiveness)])

# largest discrepancy between full and background attack effectivenesses?

# ones which lead to large changes in expression?
#AUC changes combined with binary classification setting --> genes where attacks work are ones with gene expression close to median gene expression
# if this is not happening, its not the pertubation scheme working, but median binarization artefact.
```
